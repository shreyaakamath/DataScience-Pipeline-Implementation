```python
import magellan as mg
```

```python
# get filepaths for tables A and B. 
path_A = 'tableA.csv' # path_A is the file path where table_A.csv is stored
path_B = 'tableB.csv' # path_B is the file path where table_B.csv is stored
```

```python
A = mg.read_csv(path_A, key='id')
```

```python
B = mg.read_csv(path_B, key='id')
```

```python
len(A)
```

```python
len(B)
```

```python
ob = mg.OverlapBlocker()
C = ob.block_tables(A, B, 'title', 'title', word_level=True, overlap_size=1, rem_stop_words=True,
                    l_output_attrs=['title','director', 'year','cost','star'], 
                    r_output_attrs=['title', 'director', 'year','rotten_tomatoes','review1', 'review2', 'review3','review4','review5'])
```

```python
len(C)
```

```python
feature_table = mg.get_features_for_blocking(A, B)
```

```python
feature_table
```

```python
# Create rule-based blockerdirector_directordirector_directordirector_director_lev_levdirector_director_lev_lev
rb = mg.RuleBasedBlocker()
# Add rule : block tuples if name_name_lev(ltuple, rtuple) < 0.8
rb.add_rule(['director_director_lev(ltuple, rtuple) < 0.4'], feature_table)
```

```python
D=rb.block_candset(C)
```

```python
len(D)
```

```python
D_debug=mg.debug_blocker(A, B, D, output_size=100)
```

```python
D_debug
```

```python
from __future__ import division
import pandas as pd
import jellyfish
def dir_year_match(x,y):
    # get name attribute
    x_dir = x['director']
    y_dir = y['director']
 
    #get the year attribute
    x_year = x['year']
    y_year = y['year']
    
    #if name is missing check if year matches
    if ((pd.isnull(x_dir)==True) or (pd.isnull(y_dir) == True) ):
        if ((pd.isnull(x_year)==True) or (pd.isnull(y_year) == True)):
            return False
        #for some tuples year is mm/dd/yy
        if( "/" in str(x_year)):
            x_year_list = x_year.split("/")
            year_value = int(x_year_list[2])
            if(year_value>=0 & year_value<=15):
                year_value="20"+`year_value`
            if year_value != y_year:
                return True
            else:
                return False
        #for some tuples year is "sept 21,2015" and for some it is only "2015".split based on comma to identify and correctly match
        else :
            x_year = str(x_year).strip().split(',')[-1]
            y_year =str(y_year).strip().split(',')[-1]
            # check if last names match
            if x_year != y_year:
                return True
            else:
                return False
       
    else:
        return False
```

```python
year_block= mg.BlackBoxBlocker()
year_block.set_black_box_function(dir_year_match)
```

```python
E=year_block.block_candset(D)
```

```python
E_debug=mg.debug_blocker(A, B, E, output_size=100)
```
